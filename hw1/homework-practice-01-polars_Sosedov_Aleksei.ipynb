#### 0. Ещё пара красивых слов

Первые несколько манипуляций мы сделаем за вас, а заодно поясним, в чём тут нюанс

```python
# читаем датафрейм
initial_df = pl.read_excel("end_seminar.xlsx", engine="xlsx2csv")
print(initial_df.shape[0], "rows")
print(initial_df.estimated_size("mb"), "mb")
```

Как видим, он достаточно маленький. С такими объёмами выбор библиотеки это чисто вкусовщина. Давайте немного поднимем ставки

```python
def repeat_df(df: pl.DataFrame | pl.LazyFrame, n: int = 2) -> pl.LazyFrame:
    return (
        df.lazy()
        .select(pl.col("timestamp").unique().sort().repeat_by(n).flatten())
        .join(df.lazy(), on="timestamp", how="left")
    )
```

Для работы с датафреймами, которые в память не влезают, нам понадобится объект `pl.LazyFrame`. Он представляет из себя схему выполнения вашего запроса, который оптимизируется, бьётся на батчи и параллелится под капотом через StreamingAPI, что для нас просто идеально. Перевести объект туда можно через метод `lazy`, обратно через `collect`. Учтите, что `streaming=True` работает не со всеми операциями, применяйте с умом, это можно проверить через `pl.LazyFrame.explain(streaming=True)`

```python
repeat_n = 2

repeated_df = repeat_df(initial_df, repeat_n)
print("Schema")
print(repeated_df.explain(streaming=True, optimized=True))
```

```python
# как видим, повторы работают
repeated_df = repeated_df.collect()
assert repeated_df.shape[0] == initial_df.shape[0] * repeat_n
print(repeated_df.shape[0], "rows")
print(repeated_df.estimated_size("mb"), "mb")
```

Теперь ещё интереснее. Если бы наш датасет был не в 2 раза больше, а в 100000 раз, что бы мы делали? К счастью для нас, поларс может уместить 300кк строк всего лишь в 512Мб, и это даже не прикол, но работать с этим придётся немного иначе

```python
N = 80000

df = repeat_df(initial_df, N)
print(initial_df.estimated_size("gb") * N, "gb")
```

Поэтому даже такая тривиальная задача, как подсчёт строк становится уже не такой простой. Про это и будет дз

```python
try:
    print(df.shape[0], "rows")
except AttributeError:
    print("ой, не получилось...")

# df.collect().shape[0] # тоже можно, если влезет
# len(df.collect().to_pandas()) # а такое может крашнуть ядро, попробуйте, хехе...
```

```python
def lazy_len(df: pl.LazyFrame) -> int:
    return df.select(pl.len()).collect().item()


rows = lazy_len(df)  # намного экономнее, потому что оптимизировано под капотом
assert rows == initial_df.shape[0] * N
print(rows, "rows")
```

#### 3. [0.5 балла] Посмотрите повнимательнее на колонку 'is_first_time'.

Каково процентное соотношение ответов? Сколько из них "Нет"?

```python
df3 = df.select(pl.col("is_first_time")).collect(engine="streaming")
df3 = df3.group_by("is_first_time").agg(
    pl.len().alias('count')
)
total = df3.select(pl.col('count').sum()).item()
yes = df3.filter(pl.col('is_first_time') == "Да").select(pl.col('count')).item()
no = df3.filter(pl.col('is_first_time') == "Нет").select(pl.col('count')).item()
print(
    f"ДА: {yes / total * 100:.2f}%\n"
    f"НЕТ: {no / total * 100:.2f}%"
)
```

Посмотрите на колонку `timestamp` - верно ли, что датафрейм отсортирован по ней?

```python
timestamp_df = df.select(pl.col("timestamp"))
timestamp_df = timestamp_df.with_columns(pl.col('timestamp').sort().alias("timestamp_sorted"))
timestamp_df = timestamp_df.with_columns(pl.col('timestamp').sort(descending=True).alias("timestamp_sorted_descending"))
if timestamp_df.select((pl.col("timestamp") == pl.col("timestamp_sorted")).all()).collect(engine="streaming").item():
    print("Отсортировано по возрастанию")
elif timestamp_df.select((pl.col("timestamp") == pl.col("timestamp_sorted_descending")).all()).collect(engine="streaming").item():
    print("Отсортировано по убыванию")
else:
    print("Никак не отсортировано")
```

#### 2. [2 балла] Распределите студентов в соответствии с первым приоритетом

```python
CAPACITY = {
    "Количественные финансы": 60,
    "Промышленное программирование на Haskell": 60,
    "Рекомендательные системы": 60,
    "Глубинное обучение в обработке звука": 1000,
}
```

```python
def build_applications(df: pl.DataFrame, pref_cols: list[str], required_col: str) -> pl.DataFrame:
    p1, p2, p3 = pref_cols

    applications = (
        df.select(
            "id",
            "percentile",
            pl.col(required_col).cast(pl.Int64)
            .alias("required"),
            p1, p2, p3
        )
        .unpivot(
            index=["id", "percentile", "required"],
            on=[p1, p2, p3],
            variable_name="pref_col",
            value_name="course",
        )
        .with_columns(
            pl.when(
                pl.col("pref_col") == p1
            )
            .then(1)
            .when(
                pl.col("pref_col") == p2
            )
            .then(2)
            .otherwise(3)
            .cast(pl.Int64)
            .alias("pref_order")
        )
        .with_columns(
            pl.when(
                pl.col("pref_order") == 1
            )
            .then(1)

            .when(
                pl.col("pref_order") == 2
            )
            .then(
                pl.when(pl.col("required") == 2)
                .then(1)
                .otherwise(2)
            )
            .otherwise(
                pl.when(pl.col("required") == 2)
                .then(2)
                .otherwise(3)
            )
            .cast(pl.Int64)
            .alias("wave")
        )
        .filter(
            pl.col("course").is_not_null()
            & (pl.col("course") != "")
        )
        .sort(["id", "course", "wave", "pref_order"])
        .unique(subset=["id", "course"], keep="first")
        .select(["id", "percentile", "required", "course", "pref_order", "wave"])
    )
    return applications
```

```python
def allocate(df: pl.DataFrame, pref_cols: list[str], required_col: str, max_wave: int = 3, default_capacity: int = 30) -> pl.DataFrame:
    applications = (
        build_applications(df, pref_cols, required_col)
        .with_columns(
            pl.col("course")
            .replace_strict(CAPACITY, default=default_capacity, return_dtype=pl.Int64)
            .alias("capacity")
        )
    )

    required = df.select(
        "id",
        pl.col(required_col)
        .cast(pl.Int64)
        .alias("required"),
    )

    id_dtype = df.schema["id"]
    assignments = pl.DataFrame(
        schema=
        [
            ("id", id_dtype),
            ("course", pl.String),
            ("wave", pl.Int64),
            ("pref_order", pl.Int64)
        ]
    )

    for wave in range(1, max_wave + 1):
        taken = assignments.group_by("course").agg(
            pl.len().alias("taken")
        )
        # осталось
        capacity_left = (
            applications.select(["course", "capacity"]).unique()
            .join(taken, on="course", how="left")
            .with_columns(
                pl.col("taken").fill_null(0)
            )
            .with_columns(
                (pl.col("capacity") - pl.col("taken"))
                .alias("cap_left")
            )
            .filter(pl.col("cap_left") > 0)
            .select(["course", "cap_left"])
        )
        if capacity_left.is_empty():
            break

        # кому ещё нужны
        assigned_count = assignments.group_by("id").agg(pl.len().alias("assigned"))
        need = (
            required.join(assigned_count, on="id", how="left")
            .with_columns(pl.col("assigned").fill_null(0))
            .with_columns((pl.col("required") - pl.col("assigned")).alias("need"))
            .filter(pl.col("need") > 0)
            .select(["id", "need"])
        )
        if need.is_empty():
            break

        # конкурс
        cand = (
            applications
            .filter(pl.col("wave") == wave)
            .join(need, on="id", how="inner")
            .join(capacity_left, on="course", how="inner")
            .join(assignments.select(["id", "course"]), on=["id", "course"], how="anti")
            .sort(["course", "percentile"])
            .with_columns(
                pl.int_range(0, pl.len())
                .over("course")
                .alias("rank_in_course")
            )
            .filter(pl.col("rank_in_course") < pl.col("cap_left"))
        )

        if cand.is_empty():
            continue

        # ограничение
        cand = (
            cand
            .sort(["id", "pref_order"])
            .with_columns(
                pl.int_range(0, pl.len())
                .over("id")
                .alias("rank_in_student")
            )
            .filter(pl.col("rank_in_student") < pl.col("need"))
            .select(
                "id",
                "course",
                pl.lit(wave, dtype=pl.Int64)
                .alias("wave"),

                "pref_order"
            )
        )

        if cand.is_empty():
            continue

        assignments = pl.concat([assignments, cand], how="vertical")

    return assignments
```

```python
def to_wide(df: pl.DataFrame, assignments: pl.DataFrame, required_col: str, id_out: str = "ID") -> pl.DataFrame:
    courses = (
        assignments
        .sort(["wave", "pref_order"])
        .group_by("id")
        .agg(pl.col("course").alias("courses"))
    )

    return (
        df.select("id", pl.col(required_col))
        .join(courses, on="id", how="left")
        .with_columns(
            pl.coalesce(pl.col("courses"), pl.lit([], dtype=pl.List(pl.String)))
            .alias("courses")
        )
        .with_columns(
            pl.col("courses").list.get(0, null_on_oob=True).alias("course1"),
            pl.col("courses").list.get(1, null_on_oob=True).alias("course2"),
        )
        .select(
            pl.col("id").alias(id_out),
            "course1", "course2",
            pl.col(required_col)
        )
    )
```

```python
def finalize_result(wide: pl.DataFrame, required_col: str) -> pl.DataFrame:
    return (
        wide
        .with_columns(
            pl.col("course1").fill_null("???"),
            pl.when(
                pl.col(required_col) == 1
            )
            .then(
                pl.lit("-")
            )
            .otherwise(
                pl.col("course2").fill_null("???")
            )
            .alias("course2"),
        )
        .select("ID", "course1", "course2")
    )
```

```python
from pandas.testing import assert_frame_equal
from collections import Counter

def check_first_wave_assignments(assignments: pd.DataFrame) -> None:
    """
    assignments - pandas DataFrame с колонками ['ID', 'course1', 'course2', 'fall_course_num']. Одна строка - один студент.
    - 'ID' - id студента (не объект индекса!)
    - 'course1', 'course2' - названия курсов, назначенные студенту в первую волну.
    Если один из курсов не назначен (вне зависимости от того, должен ли был быть назначен), то None.
    - 'fall_course_num' - число осенних курсов, которые должен выбрать студент.
    """

    def course_columns_to_set_column(row: pd.DataFrame):
        result = Counter()

        for column in ['course1', 'course2']:
            if (not pd.isna(row[column]) and row[column] is not None):
                result[row[column]] += 1
        return result

    ANSWER_SOURCE = 'task_2_check.csv'

    ALTERNATIVES = {
        "cdb87f7b15495e514eb481230b44540c6b0cfeb347c991f06b1e361fcdef0c0b": [
            Counter({"Глубинное обучение для текстовых данных", "Байесовские методы в машинном обучении"}),
            Counter({"Глубинное обучение для текстовых данных", }),
        ],
        "6fa015d8846e21f4f79bac013b469a1e6e962fffd6121b4d9bcde3906275293e": [
            Counter({"Statistical Learning Theory", }),
            Counter({"Statistical Learning Theory", "Байесовские методы в машинном обучении"}),
        ]
    }

    answer_df = pd.read_csv(ANSWER_SOURCE)
    assignments = assignments.copy()
    assignments['course_set'] = assignments.apply(course_columns_to_set_column, axis=1)
    answer_df['course_set'] =  answer_df.apply(course_columns_to_set_column, axis=1)

    answer_df_common = answer_df[~answer_df['ID'].isin(ALTERNATIVES.keys())][['ID', 'course_set', 'fall_course_num']]
    assignments_common = assignments[~assignments['ID'].isin(ALTERNATIVES.keys())][['ID', 'course_set', 'fall_course_num']]

    answer_df_common = answer_df_common.set_index('ID').sort_index()
    assignments_common = assignments_common.set_index('ID').sort_index()

    assert_frame_equal(answer_df_common, assignments_common, check_dtype=False, check_names=False)

    for student in ALTERNATIVES:
        assert assignments[assignments['ID'] == student]['course_set'].isin(ALTERNATIVES[student]).iloc[0]
        assert answer_df[answer_df['ID'] == student]['fall_course_num'].iloc[0] == assignments[assignments['ID'] == student]['fall_course_num'].iloc[0]

check_first_wave_assignments(assignments_pd)
```
